```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://pokemondb.net/pokedex/all'
response = requests.get(url)
html_content = response.text
soup = BeautifulSoup(html_content, 'html.parser')
```

```python
umbrellas = ['burmy', 'eevee', 'pikachu', 'tatsugiri', 'basculin', 'keldeo', 'rockruff', 'squawkabilly', 'ogerpon', 'morpeko', 'eternatus', 'greninja', 'castform']
```

```python
pokemon = {}
for row in soup.find_all('tr')[1:]:
    cells = row.find_all('td')
    name = cells[1].find('a').text.strip()
    form = cells[1].find('small')
    form = form.text.strip() if form is not None else ''
    if form == '' or name.lower() in umbrellas:
        full = name
    elif name.lower() in form.lower():
        full = form
    else:
        full = name + ' ' + form

    stats = [int(stat.text) for stat in cells[4:]]
    pokemon[full] = stats
```

```python
df = pd.DataFrame(pokemon).transpose()
df.columns = ['HP', 'Attack', 'Defense', 'Special Attack', 'Special Defense', 'Speed']
df.to_csv('stats.csv', index=True, index_label='Name')
```
